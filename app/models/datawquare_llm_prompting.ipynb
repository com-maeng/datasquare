```python
PROMPT = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
Fill out the 'Request Template' based on the following input with user request and relevant datasets. The 'Request Template' is sent from the user to the relevant datasets ower.

### Input:
User Request: {user_query}

Relevant Dataset: {dataset_1}

'Request Template'
To: (format: Dataset owner)
From: (format: User name, Team name)
Subject:
Request detail:
Data usage purpose:
Data file format (eg. .csv, .jpeg, .py, .json, etc.):
Data period: (format: YYYY-MM-DD)
Data example:

### Response:
"""
```

```python
def generate_prompt(user_query: str, dataset_1: str = None, k=0) -> str:
        return PROMPT.format(user_query=user_query, dataset_1=dataset_1)
```

```python
model = llama2_model

query = "I'm Jessica, a senior data engineer working on enhancing the performance of our recommendation system. Performance monitoring team is developing a new algorithm that aims to improve the accuracy and relevance of content suggestions for our users. To train and validate the model, I require access to detailed user interaction data, including click-through rates, viewing durations, and user feedback on recommended content from the past two years. This data is critical for fine-tuning our algorithm and ensuring it delivers personalized and engaging content to our users."
dataset = "'User feedback on recommended content' owned by 'Customer Service Team'"
# Tokenize the input text
prompt = generate_prompt(query, dataset)
inputs = tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

# Generate text
outputs = model.generate(**inputs, max_length=600, num_return_sequences=1)

# Decode the generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=False)

print(generated_text)
```
